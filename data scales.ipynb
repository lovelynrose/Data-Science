```python
education_order = ['High School', 'Bachelor', 'Master', 'PhD']
df = pd.DataFrame({'Education': ['Bachelor', 'PhD', 'Master']})

# Convert to Categorical type with specific order
df['Education'] = pd.Categorical(df['Education'], categories=education_order, ordered=True)

# Convert to integer codes
df['Education'].cat.codes
```

```python
df = pd.DataFrame({'Temperature': [20, 22, 21]})
mean = df['Temperature'].mean()
std_dev = df['Temperature'].std()
variance = df['Temperature'].var()
add = df['Temperature'] + 5  # Add 5 to each temperature value
subtract = df['Temperature'] - 2  # Subtract 2 from each temperature value
diff = df['Temperature'].diff()  # Find the difference between successive temperatures
max_temp = df['Temperature'].max()
min_temp = df['Temperature'].min()
quantile_temp = df['Temperature'].quantile(0.5)  # Median

print(f"""Mean = {mean}\nStandard Deviation = {std_dev}\nVariance = {variance}\n\
Difference = \n{diff}\nMaximum temperature = {max_temp}\nMinimum temperature = {min_temp}\n\
Median = {quantile_temp}""")
```
