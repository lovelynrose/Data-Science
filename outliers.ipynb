Finding IQR with np.percentile

```python
import numpy as np

def identify_outliers_iqr(data):
    Q1, Q3 = np.percentile(data, [25, 75])
    IQR = Q3 - Q1
    print(f"Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}")
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    print(f"lower bound = {lower_bound}, upper bound = {upper_bound}")
    outliers = [x for x in data if x < lower_bound or x > upper_bound]
    return outliers

# Example usage
data = [4, 8, 15, 16, 23, 42, 108]
outliers = identify_outliers_iqr(data)
print("Identified outliers:", outliers)
```

Finding IQR with manual method

```python
import numpy as np

def manual_quartiles(data):
    n = len(data)
    sorted_data = np.sort(data)

    # Position calculation (1-indexed)
    Q1_pos = (n + 1) / 4
    Q3_pos = 3 * (n + 1) / 4

    # Find quartiles without interpolation
    Q1 = sorted_data[int(Q1_pos) - 1] if Q1_pos.is_integer() else sorted_data[int(Q1_pos) - 1]
    Q3 = sorted_data[int(Q3_pos) - 1] if Q3_pos.is_integer() else sorted_data[int(Q3_pos) - 1]

    return Q1, Q3

# Example data
data = np.array([4, 8, 15, 16, 23, 42, 108])

# Get Q1 and Q3 using the manual method
Q1, Q3 = manual_quartiles(data)
print("Q1:", Q1, "Q3:", Q3, "IQR:", Q3-Q1)
```

Finding IQR with in-built Functions

```python
import numpy as np
from scipy import stats

# Example data
data = np.array([4, 8, 15, 16, 23, 42, 108])

# Calculating IQR
IQR = stats.iqr(data)
print("IQR:", IQR)

Q1, Q3 = np.percentile(data, [25, 75])
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = [x for x in data if x < lower_bound or x > upper_bound]
print("Outliers : ", outliers)
```

Finding z-score with in-built functions

```python
import numpy as np
from scipy import stats

# Example data
data = np.array([4, 8, 15, 16, 23, 42, 108])

# Calculating Z-scores
z_scores = stats.zscore(data)

# Identifying outliers
outliers = data[(z_scores < -2) | (z_scores > 2)]
print("Z-scores:", z_scores)
print("Outliers:", outliers)
```
